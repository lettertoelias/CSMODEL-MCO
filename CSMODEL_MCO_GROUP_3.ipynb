2. **Age** : user's age
3. **Gender** : user's gender (includes Male, Female and Non-Binary)
4. **Platform** : what social media platform the user uses
5. **Daily_Usage_Time** : Time spent on the platform daily (measured in minutes)
6. **Posts_Per_day** : number of posts the user makes per day

8. **Comments_Received_Per_day** : number of comments the user gets per day
9. **Messages_Sent_Per_day** : number of messages the user sends per day
10. **Dominant_Emotion** : the "dominant" emotional state of the user throughout the day (Happiness, Sadness, Anger, etc)
